```python
import pandas as pd

df = pd.read_excel('AU_Car_Recalls.xlsx')
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
# Hapus data penting yang kosong
df = df.dropna(subset=['Date', 'Car_ID', 'System_Affected'])

# Rapikan text
df['System_Affected'] = df['System_Affected'].str.upper().str.strip()
df['Model'] = df['Model'].str.upper().str.strip()

# Isi unit kosong
df['Units'] = df['Units'].fillna(0)

# Convert tanggal
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
```

```python
summary = df.groupby(['Model', 'System_Affected']).agg(
    total_unit=('Units', 'sum'),
    total_case=('Car_ID', 'count')
).reset_index()

summary
```

```python
df.to_csv('clean_data.csv', index=False)
summary.to_csv('summary.csv', index=False)

from google.colab import files
files.download('clean_data.csv')
files.download('summary.csv')
```
